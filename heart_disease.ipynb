```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression,LinearRegression
import seaborn as sns
from sklearn.metrics import classification_report
```

```python
cols = ['age','sex','chpain','bp','chol','FBS','EKG','HR','Exang','Stdep','sloSt','Vesfluro','Thallium','HDisease']
df = pd.read_csv('Heart_Disease_Prediction.csv')
df.columns = cols
```

```python
df["HDisease"] = (df["HDisease"]=="Presence").astype(int)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
sns.pairplot(df,hue='HDisease')
plt.show()
```

```python
for label in cols[:-1]:
    plt.hist(df[df['HDisease']==1][label],color='blue',label='Presence',density=True)
    plt.hist(df[df['HDisease']==0][label],color='red',label='Absence',density=True)
    plt.title(label)
    plt.xlabel(label)
    plt.ylabel('Count')
    plt.legend()
    plt.show()
```

```python
from pandas_profiling import ProfileReport
report = ProfileReport(df)
report.to_file("heart-disease-profile-report.html")
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Logistic Regression**

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(y_pred)
print(np.array(y_test))
```

```python
print(classification_report(y_test,y_pred))
```

**Linear Regression**

```python
linr_model = LinearRegression()
```

```python
linr_model.fit(X_train,y_train)
```

```python
y_pred = linr_model.predict(X_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
print(classification_report(y_test,y_pred))
```

**KNN Model**

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
```

```python
knn_model.fit(X_train,y_train)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```
